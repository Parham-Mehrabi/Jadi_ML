# Hierarchial  

this method doesn't require an specific number of clusters, in the end you can cut the dendrogram where you desire  
<sup><sub>a dendrogram is a branching diagram that represents the relationships of similarity among a group of entities</sub></sup>
<br/><br/><br/>

Applications:  
- Bioinformatics: grouping animals according to their biological features to reconstruct phylogeny trees
- Business: dividing customers into segments or forming a hierarchy of employees based on salary.
- Image processing: grouping handwritten characters in text recognition based on the similarity of the character shapes.
- Information Retrieval: categorizing search results based on the query.

<br/><br/><br/>


__there is two ways to do it:__
1. Divisive:
    - Divisive is top down, so you start with all observations in a single large cluster and break it down into smaller pieces
2. Agglomerative:
    - Agglomerative is opposite of divisive, so it is bottom up, where each observation starts in its own cluster and pair of clusters are merged together as they move up the hierarchy
- <small>Agglomerative is more commonAdd your tiny text</small>

##### time to create distance matrix so we can compare the pairs together:
to create a distance matrix,  
- The values of the matrix are calculated by applying a distance function between each pair of objects
- The Euclidean distance function is commonly used for this operation.

<br/>

### Similarity between Clusters
we should be able to calculate the distances between data points, but also we need to calculate the distance between clusters
__but how we compare them when they are in multi-member clusters?__

#### there are some different method to do so:
- Single Linkage Clustering   -> Minimum distance
- Complete Linkage Clustering -> Maximum distance
- Average Linkage Clustering  -> average distance from each point to all other points
- Centroid Linkage Clustering -> centroids of clusters


<br/>

#### pros:
- works with unknown N (no need to pre-specify the number of clusters)
- Easy to implement
- Useful dentograms; good for understanding
#### cons:
- impossible to undo via algorithm
- long runtime
- sometimes difficult to identify the number of clusters (specially for large datasets)


#### Hierarchical vs K-means:
- can be slower
- Doesn't require the number of clusters
- give more than one partitioning
- always generate the same clusters (there is no local best in hierarchical)
