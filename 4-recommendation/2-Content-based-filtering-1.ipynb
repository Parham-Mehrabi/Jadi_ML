# Content Based Recommenders


- works based on user Profile
- Works with user rating (like, view, ...) and then finds the similarity between the content of those content (tags, category, genres)

4. in the end to create __User's Profile__:
    - first we sum up all the scores for each genre
    - then we normalize the scores by dividing them to sum of all scores
    - for example sum of all scores would be 60(2 + 2 + 10 + 10 + 10 + 10 + 8 + 8) in this example and sum of comedy is 18(10 + 8) so comedy is 18/60 = 0.3
