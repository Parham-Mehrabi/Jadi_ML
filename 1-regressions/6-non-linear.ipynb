# Non-Linear Regression


If the data shows a curvy trend, then linear regression will not produce very accurate results when compared to a non-linear regression since linear regression presumes that the data is linear.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### this is an example for a linear data where our model is $ y = 2X+5 $

```python
x = np.arange(-5.0, 5.0, 0.1)
y = 2*(x) + 5
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.show()
```

### this is an example for a non-linear data where our model is $ y = 2x^3 + 2x^2 + 3x +10 $

```python
x = np.arange(-5.0, 5.0, 0.1)
y = 2*(x**3) + 2*(x**2) + 3*x + 10
y_noise = 20 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.show()
```

### Non-Linear functions can have elements like:
- exponential
- Logarithmic
- Sigmoidal/Logistic/S-curve
- fraction
- Quadratic
- etc
#### they can be either simple like $ y = \log(x) $ or complex like $ y = log(ax^3 + bx^2 + cx + d) $

## Quadratic: 
#### $ y = x^2 $

```python
x = np.arange(-5.0, 5.0, 0.1)

y = np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.show()
```

## Exponential:
##### $ y = a + bc^x $
where b ≠0, c > 0 , c ≠1, and x is any real number. The base, c, is constant and the exponent, x, is a variable.
#### $ y = a + bc^x | b\neq{0}, c\neq{1}, c>1, x\in{R}$




```python
x = np.arange(-5, 5, 0.1)
y = np.exp(x)
plt.plot(x, y)
plt.show()
```

# Logarithmic:
$ y = \log{X} $

X can be a polynomial representation of the $ x $ values.

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = np.log(X)
plt.plot(X,Y) 
plt.show()
```

# Sigmoidal/Logistic/S-curve:
$ y = a + \frac{b}{1 + c^(X-d) } $ 

```python
X = np.arange(-5.0, 5.0, 0.1)

Y = 1 - (4 / (1 + np.power(3, X-2)))
plt.plot(X,Y) 
plt.show()
```
